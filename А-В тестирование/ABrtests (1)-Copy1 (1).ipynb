### Приоритизация гипотез

```python
# загрузим данные
try:
    data_g = pd.read_csv(r'C:\Users\User\Downloads\hypothesis.csv')
except:
    data_g = pd.read_csv(r'https://code.s3.yandex.net/datasets/hypothesis.csv')
```

```python
# посмотрим на таблицу гипотез
pd.set_option('max_colwidth', 200)
data_g
```

В данных имеется список гипотез для увеличения продаж, проведем приоритизацию методом ICE

```python
# добавим столбец с расчетами приоритизации, отсортируем гипотезы по убыванию оценки приоритизации
data_g['ICE'] = ((data_g['Impact']*data_g['Confidence'])/data_g['Efforts']).map(lambda x: round(x,2))
data_g = data_g.sort_values(by = 'ICE', ascending = False)
```

```python
# отобразим результат
pd.set_option('max_colwidth', 200)
data_g
```

Согласно приоритизации гипотез методом ICE в первую тройку входят гипортезы 8, 0, 7 , а именно "Запустить акцию, дающую скидку на товар в день рождения", 'Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей','Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок'.Самой неприоритетной считается 4 гипотеза, предлагающая поменять цвет фона главной страницы. Приоритизация выглядит вполне логичной с точки зрения цели потсановки гипотез, а имеено увеличение продаж. Произведем расчет приоритизации гипотез мнтодом RICE, добавив в числитель формулы оценку обьема пользователей, которых затронет изменение.

```python
# добавляем столбец расчетом RICE, отсортируем по убыванию значения оценки
data_g['RICE'] = ((data_g['Impact']*data_g['Confidence']*data_g['Reach'])/data_g['Efforts']).map(lambda x: round(x,2))
data_g = data_g.sort_values(by = 'RICE', ascending = False)
```

```python
# отобразим результат
pd.set_option('max_colwidth', 200)
data_g
```

При учете охвата аудитории, приоритизация гипотез изменилась, первая тройка - это гипотезы 7, 2, 0, а именно "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок", 'Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа', 'Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей'. При среднем уровне затрат, решающую роль сыграло значение охвата аудитории, у гипотезы 7, максимальный показатель, равный 10, так как при методе RICE в числитель дроби добавляется именно охват пользователей. У гипотезы 8 этот показатель равен 1, так как скидкка дается только 1 раз в году - охват пользователей скорее всего будет небольшой.Изменение фона главной страницы все также имеет самый низкий приоритет при использовании модификации RICE. Так как 7 гипотеза имеет высокую оценку охвата пользователей, влияние на пользователей,уверенность в изменениях и при этом средний уровень затрат, рекомендуется тестировать 7 гипотезу первой.

Загрузим данные об общем количестве покупателей за период наблюдения.

```python

# загрузим данные
try:
    visitors = pd.read_csv(r'C:\Users\User\Downloads\visitors.csv')
except:
    visitors = pd.read_csv(r'https://code.s3.yandex.net/datasets/visitors.csv')
```

```python
# проверим данные на явные дубликаты, 0 - дубликаты отсутствуют
visitors.duplicated().sum()
```

```python
# переведем столбец с датой в нужный формат
visitors['date'] = visitors['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
```

```python
# проверим сходятся ли даты эксперимента
print('Дата начала эксперимента: ', visitors['date'].min())
print('Дата последнего наблюдения: ', visitors['date'].max())
```

Данный о количестве визитов содержат дату, группу наблюдения и общее количество посетителей в выбранную дату. Информация о периоде наблюдения совпадает.Произведем расчеты

```python
# создадим таблицу, сгруппировав заказы по группе и дате, посчитаем общую выручку в день и количество покупателей
ordersCum= orders.groupby(['group','date'], as_index = False).agg({'revenue': 'sum', 'visitorId':'count'})

# создадим столбец с кумулятивной выручкой отдельно для каждой группы
ordersCum['cumulative_rev'] = ordersCum.sort_values(by = ['group','date']).groupby('group')['revenue'].cumsum(axis = 0)

# создадим столбец с кумулятивным количеством заказов
ordersCum['cumulative_orders'] = ordersCum.sort_values(by = ['group','date']).groupby('group')['visitorId'].cumsum(axis = 0)

# присоединим к таблице данный об общем количестве посетителей на дату в каждой группе
ordersCum = ordersCum.merge(visitors.groupby(['group','date'], as_index = False).agg({'visitors': 'sum'}),
                           left_on = ['group','date'], right_on = ['group','date'])
# создадим столбец с кумулятивным количеством посетителей на дату в каждой группе
ordersCum['cumulative_visitors'] = ordersCum.sort_values(by = ['group','date']).groupby('group')['visitors'].cumsum(axis = 0)

# посчитаем конверсию - среднее количество заказов на посетителя
ordersCum['convers'] = ordersCum['cumulative_orders']/ordersCum['cumulative_visitors']
# переименуем столбцы для удобства
ordersCum.columns = ['group','date', 'revenue', 'buyer_count','cumulative_rev','cumulative_orders','visit_count',\
                    'cumulative_visitors', 'convers'] 
```

```python
# проверим сходятся ли даты эксперимента
print('Дата начала эксперимента в сводном датасете: ', ordersCum['date'].min())
print('Дата последнего наблюдения в сводном датасете: ', ordersCum['date'].max())
```

Таблица `ordersCum` содержит информацию о группе наблюдения, дате события, выручке в дату, количестве покупателей, накопительную выручку, накопительное количество заказов, общее количество посетителей, накопительное количество посетителей и конверсию(среднее количество заказо на посетителя).

#### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.

Смоделируем набор данных, отражающий количество заказов для каждого посетителя магазина для каждой группы наблюдения отдельно. Значение уровня значимости примем за 0,05. Нулевая гиотеза подразумевает утверждение, что статистически значимых различий между наборами данных по группе А и В нет. Альтернативная гипотеза имеет противоположный смысл - статистически значимые разлия в среднем количестве заказов между группами есть. Так как данные связаны с исследованием поведенческих характеристик пользователя, а именно число заказов и сумма заказов, можно предположить, что наборы данных будут иметь ненормальное распределение, выбросы и анамолии. Для повышения качества исследования применим непараметрический тест Манна-Уитни

```python
# набор данных для группы A
sampleA = pd.concat([orderperuser[orderperuser['group']=='A']['transactionId']\
,pd.Series(0, index=np.arange(ordersCum[ordersCum['group']=='A']['cumulative_visitors'].iat[-1]\
                              - len(orderperuser[orderperuser['group']=='A'])), name='orders')],axis=0)
```

```python
# набор данных для группы В
sampleB = pd.concat([orderperuser[orderperuser['group']=='B']['transactionId']\
,pd.Series(0, index=np.arange(ordersCum[ordersCum['group']=='B']['cumulative_visitors'].iat[-1]\
                              - len(orderperuser[orderperuser['group']=='B'])), name='orders')],axis=0)
```

```python
print('p-value =','{0:.3f}'.format(stats.mannwhitneyu(sampleA,sampleB)[1]))
```

```python
print('Относительный прирост среднего числа заказов группы B: ','{0:.3f}'\
                                                              .format(sampleB.mean()/sampleA.mean() - 1))
```

Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным равна 0,017, что значительно меньше принятого уровня значимости. Можно отвергнуть нулевую гипотезу и предположить, что вероятность получить такие или еще более выраженные различия между группами, при условии, что нулевая гипотеза верна, крайне мала. Между наборами 'сырых' данных, отражающих среднее количество заказов на посетителя по группам наблюдения есть сатитистически значимые отличия. Группа В лидирует с приростом 13,8%

#### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным

Смоделируем набор данных, отражающий количество заказов для каждого посетителя магазина для каждой группы наблюдения отдельно. Значение уровня значимости примем за 0,05. Нулевая гиотеза подразумевает утверждение, что статистически значимых различий между наборами данных по группе А и В нет. Альтернативная гипотеза имеет противоположный смысл - статистически значимые разлия в среднем количестве заказов между группами есть.

```python
# создадим список пользователей с заказами больше 2
abnormal_orders = list(orderperuser[orderperuser['transactionId']>\
                                    np.percentile(orderperuser['transactionId'],[95])[0]]['visitorId'].unique())
```

```python
# создадим список пользователей с заказами больше 28000
abnormal_rev = list(orders[orders['revenue']>np.percentile(orders['revenue'],[95])[0]]['visitorId'].unique())
```

```python
# создадим список с уникальными номерами пользователей с заказами больше 2 и суммой заказа больше 28000
abnormal_user = list(set(abnormal_orders + abnormal_rev))
```

```python
# набор отфильтрованных данных для группы A
sampleAfilter = pd.concat([orderperuser[(orderperuser['group']=='A')&\
                                  (~orderperuser['visitorId'].isin(abnormal_user))]['transactionId']
,pd.Series(0, index=np.arange(ordersCum[ordersCum['group']=='A']['cumulative_visitors'].iat[-1]\
                              - len(orderperuser[orderperuser['group']=='A'])), name='orders')],axis=0)
```

```python
# набор отфильтрованных данных для группы B
sampleBfilter = pd.concat([orderperuser[(orderperuser['group']=='B')&\
                                  (~orderperuser['visitorId'].isin(abnormal_user))]['transactionId']
,pd.Series(0, index=np.arange(ordersCum[ordersCum['group']=='B']['cumulative_visitors'].iat[-1]\
                              - len(orderperuser[orderperuser['group']=='B'])), name='orders')],axis=0)
```

```python
print('p-value =','{0:.5f}'.format(stats.mannwhitneyu(sampleAfilter,sampleBfilter)[1]))
```

```python
print('Относительный прирост среднего числа заказов группы B: ','{0:.3f}'\
                                                              .format(sampleBfilter.mean()/sampleAfilter.mean() - 1))
```

Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным равна 0,013, что значительно меньше принятого уровня значимости.Можно отвергнуть нулевую гипотезу и предположить, что между группами А и В имеется статистически значимые различия. Причем относительный выйгрыш группы В равен 17%

#### Cтатистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

Значение уровня значимости примем за 0,05. Нулевая гиотеза подразумевает утверждение, что статистически значимых различий между наборами данных по группе А и В нет. Альтернативная гипотеза имеет противоположный смысл - статистически значимые разлия в среднем чеке заказа между группами есть.

```python
print('p-value =','{0:.3f}'.format(stats.mannwhitneyu(orders[(orders['group']=='A')&\
                                  (~orders['visitorId'].isin(abnormal_user))]['revenue'],
                                         orders[(orders['group']=='B')&\
                                  (~orders['visitorId'].isin(abnormal_user))]['revenue'])[1]))
```

Значение p-value значительно превышает принятый уровень статистической значимости, отвергнуть нулевую гипотезу нельзя, а значит можно предположить, что статистически значимых различий в среднем чеке между группами нет.  Проверим относительный прирост группы В по отношению к группе A.

```python
print('Относительный прирост среднего чека заказов группы B: ','{0:.3f}'.format(orders[(orders['group']=='B')&\
                                  (~orders['visitorId'].isin(abnormal_user))]['revenue'].mean()/
                                         orders[(orders['group']=='A')&\
                                  (~orders['visitorId'].isin(abnormal_user))]['revenue'].mean() - 1))
```

Относительный прирост среднего чека заказов группы B проседает по отношению к группе В на 2% после удаления аномальных стоимостей заказов
