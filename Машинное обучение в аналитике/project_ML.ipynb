### Построение модели прогнозирования оттока клиентов

#### Логистическая регрессия

Построим модель прогнозирования оттока методом Логистической регресии

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = data.drop(columns=['Churn'])
y = data['Churn']
```

```python
# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)# обучите scaler на обучающей выборке методом fit
X_train_st =scaler.transform(X_train)# стандартизируйте обучающую выборку методом transform scaler 
```

```python
# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

```python
model = LogisticRegression(solver='liblinear')
model.fit(X_train_st,y_train)
```

```python
# бинарный прогноз
predictions = model.predict(X_test_st)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
probabilities = model.predict_proba(X_test_st)[:, 1]
```

```python
# вывем метрики для полученного прогноза
print('Accuracy: {:.2f}'.format(accuracy_score(y_test,predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test,predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test,predictions)))
print('F1: {:.2f}'.format(f1_score(y_test,predictions)))
print('ROC_AUC: {:.2f}'.format(roc_auc_score(y_test, probabilities)))
```

Метрика Accuracy - доля правильных ответов находится на довольно высоком уровне - 90%. Высокий показатель можно обьяснить довольно смещенным балансом классов. `Precision` - доля правильных ответов только среди целевого класса '1' равна 83%. Метрика имеет важное значение, так как факт отнесения к оттоку при противоположном реальном решении может стоить дополнительных потраченных в пустую маркетинговых ресурсов. Например предоставление скидки на абонемент, при ситуации, когда пользователь продолжил бы посещение без дополнительной стимуляции.`Recall` на уровне 78%, т.е среди всех обьектов класса '1' в реальности обнаружены 78%. Показатель находится также на достаточно высоком уровне. Баланс , а именно метрика `F1`, показывает хороший результат на уровне 81% .Метрика оценки качества классификатора показывает значение в 95%, что очень близко к 100%.Так как доля класса '0' в общем обьеме в 3 раза превышает класс '1' скорректируем значение уверенности к отнесению конкретного класса, уменьшим порог на 0,05 для балансировки метрик

```python
# задаем порог
threshold = 0.45

# на основании вероятностей и соотношения классов рассчитаем predict
custom_predictions = [0 if i < threshold else 1 for i in probabilities]

# метрики для прогноза по новому порогу
print('Метрики для прогноза с кастомным порогом')
print(
    'Accuracy for custom: {:.2f}'.format(
       accuracy_score(y_test, custom_predictions)
    )
)
print(
    'Precision for custom: {:.2f}'.format(
       precision_score(y_test, custom_predictions)
    )
)
print(
    'Recall for custom: {:.2f}'.format(
        recall_score(y_test, custom_predictions)
    )
)
print('F1 for custom: {:.2f}'.format(f1_score(y_test, custom_predictions)))
```

На основе скорректированного порога удалось при минимальной потере Precision на 2 п.п, увеличить метрику Recall на 5 п.п и баланс F1 на 1 п.п
