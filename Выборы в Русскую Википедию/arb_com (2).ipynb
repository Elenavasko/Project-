### Результирующие итоги

Для дальнейшего подсчета голосов и определении победителей, голоса с пометкой недействующих учитываться не будут.

```python
# создадим датасет для подсчета голосов, отфильтруем действующие голоса
data_true = data.query('can_vote == True')
# выделим текущую дату
data_true['dt'] = data_true['time'].dt.date
```

```python
# функция для получения результата выборов на текущую дату выбооров
def get_percent(data):
    
    # для каждой компании выборов
    for n in data['n'].unique():
        # создадим сводную таблицу для подсчета результирующего итога на каждую дату выборов
        df = data.query('n ==@n and can_vote == True').pivot_table(index = ['candidate','dt']\
          ,values = 'vote',aggfunc =['count',lambda x:(x == 1)\
         .sum()]).rename(columns = {'count':'total_voice','<lambda>':'positive'})
        df.columns = df.columns.droplevel(1)

        df['total_voice'] = df.sort_values(by = ['candidate','dt']).groupby('candidate')['total_voice'].cumsum(axis = 0)
        df['positive'] = df.groupby('candidate')['positive'].cumsum(axis = 0)
        df['percent'] = df.apply(lambda x: round(x['positive']/x['total_voice']*100,2), axis = 1)
        df1 = df.pivot_table(index = ['candidate'], columns = 'dt', values = 'percent')
        df1 = df1.T.reset_index()
        # посторим линейный график 
        df1.plot(kind = 'line', x = 'dt', figsize = (8,4), \
                                       title= f'Результирующий итог на текущую дату в {n} выборах')
        # отметим пороговое значение в 66% для победы
        plt.axhline(y = 66, linestyle='dotted', linewidth=2 ,label='Пороговое значение')
        plt.ylabel('Результирующий процент')
        plt.xlabel('Дата выборов')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()
```

```python
get_percent(data_true)
```

Можно заметить, что общая тенденция обозначается в перый день выборов.Результирующий итог подвержен временному влиянию, но незначительно. Особое внимание можно обратить на пороговые значения первого дня выборов от 60% до 75 %. В этом диапазоне могут возникать перемены прогноза, так как значения находятся ближе к победному отметке в 66 %

Определим есть ли пересекающиеся кандидаты на разных выборах и есть ли постоянные победители

```python
# создадим сводную таблицу, выделим победителей
df_ret = data_true.pivot_table(index = ['n','candidate'],values = 'vote', \
     aggfunc =['sum','count',lambda x:(x == 1).sum()]).\
rename(columns = {'sum':'result','count':'total_voice','<lambda>':'positive'})
df_ret.columns = df_ret.columns.droplevel(1)
df_ret['win'] = df_ret.apply(lambda x: 1 if round(x['positive']/x['total_voice']*100,2) >=66.66 else 0, axis = 1)
```

```python
#создадим словарь, где ключами будут ники кандидатов, а значения это количество вхождений в разные кампании выборов
dict = {}
for n in df_ret.reset_index()['n'].unique():   
    for uni_name in df_ret.reset_index().query('n == @n')['candidate'].unique() :
        dict[uni_name] = dict.get(uni_name,0)+1 
# создадим список кандидатов, которые участвовали в 2 и более выблорах
result = []
for key, value in dict.items():
    if value>=2:
        result.append(key)      
```

```python
# отфильтруем данные по нику кандидата
df_ret1 = df_ret.reset_index().query('candidate in @result')[['n','candidate','win']]
# построим график участия в выборах с указанием результата голосования
sns.catplot(x = 'n', y = 'candidate', hue = 'win', data = df_ret1, kind = 'strip',legend=False)
plt.title('Кандидаты участвующие в разных кампаниях 2 и более раз')
plt.xlabel('Номер выборов')
plt.ylabel('Ник кандидата')
plt.legend(title = 'Статус победителя', bbox_to_anchor=(1, 0.5) );
```

Некоторые кандидаты участвую в череде выборов и имеют по 4 ненудачи, после неудачной попытким может вполне быть победный тур или наоборот. Абсолютных фаворитов нет, можно говорить о сменяемости предпочтений от кампании к кампании.Стоит отметить , что самы многочисленный возврат случился на 32 выборах, 10 кандидатов уже имели опыт предыдущих выдвижений, еще 6 предпримут попытки в дальнейшем 

### Оппортунист/Конформист

Просегментируем пользователей на конформистов и оппортунистов

```python
# создадим датасет для сегментаци
data_merge = data_true.merge(df_ret.reset_index()[['n','candidate', 'win']], on = ['n', 'candidate'], how = 'left')
data_merge.head(5)
```

```python
# так как сумма голосов и результата кандидата для победы может иметь значение -1(против(-1) + проигрыш(0)) или 
# 2 (победа (1) и за (1)) создадим столбец с флагом результата, 1 = конформист, 0 -оппортунист
data_merge['flag'] = data_merge.apply(lambda x: 1 if (x['win'] + x['vote']) in [-1,2] else 0, axis = 1)
# посчитаем в процентном итоге напсколько избиратель предсказал исход
data_merge['op'] = data_merge.groupby(['voter','n'])['flag'].transform('sum')\
                       /data_merge.groupby(['voter','n'])['flag'].transform('count')
# создадим столбец с флагом конформиста, где результат отдельного голосующего на конкретных выборах больше 0,5
data_merge['is_conformist'] = data_merge.apply(lambda x: 1 if x['op']> 0.5 else 0, axis = 1)   
```

```python
# изобразим результат на графике
data_conform = data_merge.drop_duplicates(subset = ['voter','n','is_conformist']).pivot_table(index = ['n'], \
                                        values = 'is_conformist', aggfunc =['sum', lambda x:(x == 0).sum()])
data_conform.columns = data_conform.columns.droplevel(1)
data_conform.rename(columns = {'sum':'conformist','<lambda>':'opportunist'}).plot(kind = 'bar', figsize =(8,4))
plt.title('Количество оппортунистов и комформистов в кампаниях')
plt.ylabel('Количество голосующих')
plt.xlabel('Номер выборов');
```

Можно с увенностью говорить, что конформистов большинство, что может обьясняться желанием голосовать за потенциального победителя и отметать потенциального проигравшего.Рассмотрим динамику простановки голосов во времени для визуализации активности голосующих

```python
# напишем функцию для визуализации временной динамики голосов оппортунистов и конформистов
def hour_conf(data):
    # для каждых выборов
    for n in data['n'].unique():
        # посторим динамику распределение голосов во времени
        g = sns.catplot(x = 'time',data = data.query('n ==@n'),col = 'is_conformist',  kind='strip', \
                        alpha = 0.4,height = 3, aspect = 1.5)
        # отформатируем графики
        g.set_xticklabels(rotation = 90,fontsize=9)
        g.set_xlabels('Длитильность выборов',fontsize=9)
        axes = g.axes.flatten()
        axes[0].set_title(f'Динамика голосования оппортунистов, {n} выборы',fontsize=10)
        axes[1].set_title(f'Динамика голосования конформистов, {n} выборы',fontsize=10)
```

```python
# визуализируем данные
hour_conf(data_merge)
```

Активнее всего голосование у конформистов проходит  в первые сутки,далее количество резко снижается и распределяется относительно равномерно до последнего дня голосования.В 33 и 34 выборах есть небольшое увеличение количества голосов в последний день выборов. Отличается поведение постановки голосов в 32 выборах: плотная постановка голосов в первые 2 суток голосования и последние часы голосования последнего дня. Если учесть, что именно эти выборы отличаются повышенной активностью голосующих оносительно кандидатов можно предположить некоторые скрытые причины повышения конкуренции.Оппортунисты в целом имеют схожую динамику.31 выборы имеют наименьшую активность, 33 наибольшую .Первый и последний дни голосования как правило набирают количественно больше голосов. Если учесть динамику результирующего процента и постановки голосов, то можно сказать, что если голосующий воздержится от постановки голоса в первые 2 суток, то вполне может проголосовать как 100% конформист

Содержимое столбца `H/d` имеет строковый тип, переведем среднее время работы в минуты, числовой формат. 

```python
# функция для перевода времени в нужный формат в минуты
def avg_time(string):
    if len(string) > 4:
        position = string.index('h')
        result = int(string[:position].strip())*60 + int(string[(position+1):].lstrip(' 0'))
    else:
        if 'm' in string:
            result = int(string[:-1].strip())
        else:
            result = int(string[:-1].strip())*60    
    return result
```

Среднее количество голосов отданных избирателями самой высокое в 32 выборах, с показателем около 14 голосов на 1 избирателя, что существенно отличается от общей тенденции по выборам. Голоса избирателей содержат отметку о действительности голоса, которая не может менятся в течение одних выборов. Ситуации о смене статуса голосующего возможна при отмене голоса. Таких пользоваиелей всего 9 человек, а значит избиратели редко меняют решение при выборе. Голосование предусматривает отметку за свою же кандидатуру. Таких избирателей 18 человек, причем некоторые ставят себе отметку "против".
 
 Результат подсчета победителей предполагает, что 2/3 голосов за кандидутру должны быть с отметкой "за" от общего количества голосов за этого кагдидата. Нейтральные голосам не учитываются. Кумулятивный итог по дням выборов для каждого кандидата показывает, что именно первые сутки голосования предсказывают итог, перемена прогноза может произойти только если результат первого дня варируется примерно от 60 до 75%, в этом случае может быть противоположный результат.
  
  Состав кандидатов на выборах меняется, но есть и повторы. 23 кандидата из общего числа уникальных кандидатов участвовали в выборах более 2 раз, с переменным успехом. Стоит отметить, что есть кандидаты, которые выигрывали каждые свои выборы, есть те, кто проиграл все туры. Максимальное количество выборов, в которых участвовал кандидат равняется 4,исходя из имеющихся данных.
   
   Для характеристики госолосующих произведена сегментация на конформистов и оппортунистов. Метод сегментизации предполагает, что если половина отданных избирателем голосов пришлись на выйгравших по итогу выборов кандидатов, то такой голосующий считается конформистом, если результат противоположный, то оппортунистом. Конформисты уверенно лидируют по количеству на всех выборах с макисмумом в 35 выборах,причем конформисты имеют в целом похожее распределение количества на всех выборах. Оппортунисты имеют свой максимум в 33 выборах, и ощутимо проседающий минимум в 31 выборах. Временная шкала визуализации поведения конформистров и оппортунистов во времени показывает, что и те и другие предпочитают отдавать голоса в первые сутки голосования и последний день. Так как количество кандидатов в 32 выборах заметно вывше, чем в отстальных,а 33 и 34 выборы имеют максимумы голосующих, временные ряды этих выборов показывают повышенную активность. Причем для 32 выборов имеется особенность в виде увеличения голосов не только в первые 2 суток, но и в последние часы голосования.
   
   Данные по пользовательским характеристикам содержат информацию о нике пользователя, общем количестве правок, количестве энциклопедических правок, общем количестве созданных страниц и количестве созданных энциклопедических страниц, количестве добавленногго теста, месяцев пребывания в проекте, среднему времени нахождения на сайте в мин/день, логи действий вне правок.
   Произведена сегментация по квартилям на 4 группы по общему количеству правок, общему количеству созданных страниц, среднему времени. 
   Среднее значение количества правок на 1 пользователя ожидаемого самое высокое в 4 группе по количеству правок с результатои около 70000 правок, тогда как первая группа имеет менее 5000 правок на пользователя. Тем не менее медианный показатель доли корректных правок в 4 группе один из самых низких, что может обьяснятся увеличивающися шансом на ошибку в связи с большим обьемом вносимой информации. Также стоит ометить, что с увеличением числа правок как правило и увеличивается показатель длительности нахождения пользователя на проекте, что можно считать вполнет закономерным, даже несмотря на наличие выбросов.Размах доля энцеклопедических правок снижается по мере увеличения общего количества правок, т.е от 1 категории к 4.
   В сегментах по общему количеству созданных страниц, 4 группа ожидаемо иммеет самый высокий уровень, среднее значения - это около 5000 страниц на пользователя, в то время как первая группа иммет показатель равный менее 150. Доля энциклопедических страниц выше в первой группе, медианное значение около 0,3 и имеет увеличенный размах, наименьшее медианное значение в четвертой группе, что говорит о том, что с увеличение количества создаваемых страниц, количество энциклопедических не растет. Распределение значений общего количества правок внутри категорий созданных страниц показывает, что чем меньше правок вносит пользователь, тем меньше создает страниц, что вполне ожидаемо.
   
   Введены некоторые дополнительные метрики, такие как активность пользователя(вне правок), выраженное в количестве действий, созданных пользователям в месяц и производительность, рассчитанная как количество символов в день, добавленное пользователем за все время участия проекте.
    Самыми активными пользователями в сегментах по количеству правок и созданных страниц являются пользователи 4 группы. Средняя активность в первой и четвертой группах отличаются почти в 10 раз. Средняя производительность согласно этим сегментам имеет похожее распределение, за исключением разницы, производительность первой и четвертой группы отличается в 4 раза, в пользу последней.
    В сегметах по среднему времени пребывания в день самой производительной является 4 группа, хотя в первой группе, несмотря на небольшой размах,  есть пользователи со значительным превышением показателя, оносительно общего распределения. Размах значений активности в этих сегментах говорит о том, что пользователи 4 группы имеют весьма разные показатели активности, что скорее всего связанно с большим лагом времени в этой группе (от 70 до 705 минут посещения в день)
    Среди пользователей имеющих право голоса, разделенных на сегменты по количеству созданных страниц, правок и среднему времени в каждой из групп уверенно лидируют конформисты. Больше всего голосующих в сегментах по колиеству правок в 4 группе. В сегментах по количеству проведенного времени лидирует первая группа, со среднем временем 5-30 минут. В категориях по количеству правок также лидирует первая группа. Но самая многочисленная группа по численности не является самой активной. Средние значения активности и производительности в сегментах по количеству правок и созданных страниц у избирателей имеют схожую картину с пользовательскими - уверенно лидирует 4 группа. Распределение значений доли корректных правок у голосующих во всех 4 группах имеет немного сниженное медианное значение, чем в целом по пользователям. Распределение доли энциклопедических правок для 2, 3 и 4 групп голосующих также имеет меньшее медианное значение. В категориях по количеству созданных страниц наименьшее количество кандадатов в первой группе, наибольшее в четвертой.
